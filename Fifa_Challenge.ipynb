```python
# cleaning with functions

pd.set_option('display.max_rows', 105)

def percentage_na(df):
    df2 = df.copy()
    nulls_df = pd.DataFrame((df2.isna().sum()/len(df2))*100)
    nulls_df = nulls_df.reset_index()
    nulls_df.columns = ['column','percentage']
    return nulls_df
```

```python
nulls_df = percentage_na(df)
display(nulls_df[nulls_df.percentage>0])
```

```python
nulls_df[nulls_df['percentage']>50]['column'].values.tolist()
```

```python
def drop_cols_with_too_nas(df, nulls_df, threshold=50):
    df2 = df.copy()
    cols_to_drop = nulls_df[nulls_df['percentage'] > threshold]['column'].values.tolist()
    df2 = df2.drop(cols_to_drop, axis=1)
    return df2
```

```python
df = drop_cols_with_too_nas(df, nulls_df)
```

```python
def replace_values(df, nulls_df, threshold = 2):
    df2 = df.copy()
    cols_to_replace = nulls_df[nulls_df['percentage'] > threshold]['column'].values.tolist()
    for item in cols_to_replace :
        df2[item] = df2[item].fillna('Unknown')
    return df2
```

```python
df = replace_values(df, nulls_df)
```

```python
def drop_rows(df, nulls_df, threshold = 0):
    df2 = df.copy()
    cols_rows_to_drop = nulls_df[nulls_df['percentage'] > threshold]['column'].values.tolist()
    for item in cols_rows_to_drop :
        df2 = df2[(df2[item].isna()==False)]
    return df2
```

```python
df = drop_rows(df, nulls_df)
df.shape
```

```python
def drop_duplicates(df) :
    df2 = df.copy()
    df2 = df2.drop_duplicates()
    return df2
```

```python
df = drop_duplicates(df)
df.shape
# no duplicates
```

```python
# now cleaning columns : first height

def clean_height(x):
    if ("'" in x) :
        return float(x.replace("'" , ".")[:-1])  #[:-1] removes the last character (so the double quote)
    else :
        return x
    return x
```

```python
df['height'] = df['height'].apply(clean_height)
df['height'].unique()
```

```python
# df['height'] = list(map(clean_height_2,df['height']))
# df['height'].unique()
```

```python
# Then, weight

def clean_weight(x):
    if ("lbs" in x) :
        return int(x.replace("lbs",""))
    else :
            return x
    return x
```

```python
df['weight'] = list(map(clean_weight,df['weight']))
df['weight'].unique()
```

```python
# then the dates

def clean_date(x):
    x = pd.to_datetime(x, errors='coerce')
    x = x.strftime(format='%d-%m-%Y')
    return x
```

```python
df['joined'] = list(map(clean_date,df['joined']))
df['joined'].unique()
```

```python
# then amount in euros

def clean_euros(x):
    if  ("K" in x) :
        return float(x[1:-1]) * 1000
    elif  ("M" in x) :
        return float(x[1:-1]) * 1000000
    else:
        return float(x[1:])
```

```python
df['value'] = list(map(clean_euros,df['value']))
df['value'].unique
```

```python
df['wage'] = list(map(clean_euros,df['wage']))
df['wage'].unique
```

```python
df['release clause'] = list(map(clean_euros,df['release clause']))
df['release clause'].unique
```

```python
# then the stars

def clean_star (x) :
    if ("★" in x):
        return int(x.replace("★",""))
    else :
        return x
```

```python
df['w/f'] = df['w/f'].apply(clean_star)
df['sm'] = df['sm'].apply(clean_star)
df['ir'] = df['ir'].apply(clean_star)
```

```python
df['w/f'].unique()
df['sm'].unique()
df['ir'].unique()
```

```python
# finally football positions :

def sum_num(text):
    return sum([float(x) for x in text.split('+')])
```

```python
pd.set_option('display.max_columns', 105)
```

```python
for column in ['ls','st','rs','lw','lf','cf','rf', 'rw', 'lam', 'cam', 'ram', 'lm', 'lcm', 'cm', 'rcm', 'rm', 'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb', 'lcb', 'cb', 'rcb', 'rb', 'gk']:
    df[column] = df[column].apply(sum_num)
display(df.head())
```
